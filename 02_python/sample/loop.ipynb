```
i = 0

while True:
    i += 1

    if i > 10:
        break

    print(i)

print("処理を終了します")
```

```
i = 0

while True:
    i += 1

    if i > 10:
        break

    if i % 2 == 1:
        continue

    print(i)

print("処理を終了します")
```
