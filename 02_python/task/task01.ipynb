```
def introduce(name: str, age: int) -> None:
    """
    名前と年齢を受け取り、自己紹介を表示する。

    Args:
        name: 名前。
        age: 年齢。
    """
    print(f"はじめまして、{name}と申します。")
    print(f"年齢は{age}歳です。来年は{age + 1}歳になります。")
    print("よろしくお願いします。")


if __name__ == "__main__":
    introduce("テスト太郎", 20)
```
