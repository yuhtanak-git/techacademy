```
def calculate_price_details(price_excluding_tax: int) -> tuple[int, int]:
    """
    税抜価格から税込価格と送料を計算する。

    Args:
        price_excluding_tax: 税抜価格

    Returns:
        (税込価格, 送料)
    """
    tax_included_price = int(price_excluding_tax * 1.1)
    shipping_fee = 0 if tax_included_price >= 2000 else 350
    return tax_included_price, shipping_fee


def main() -> None:
    """
    ユーザーから税抜価格を入力として受け取り、
    税込価格と送料を計算して表示する。
    """
    try:
        price = int(input("税抜価格を入力してください: "))
        tax_price, shipping_fee = calculate_price_details(price)

        if shipping_fee == 0:
            print("送料は無料です")
        else:
            print("送料は350円です")

        total_price = tax_price + shipping_fee
        print(f"送料込税込価格は{total_price}円です")

    except ValueError:
        print("数値を入力してください")
    except KeyboardInterrupt:
        print("処理を中断しました。終了します。")


if __name__ == "__main__":
    main()
```
