```
def analyze_grades(grades: list[int]) -> tuple[float, int, int, int]:
    """
    成績リストを分析し、平均・最高点・最低点・平均以上人数を返す。

    Args:
        grades: 学生の成績リスト。

    Returns:
        平均点, 最高点, 最低点, 平均以上の学生数
    """
    if not grades:
        raise ValueError("成績リストが空です")

    average = sum(grades) / len(grades)
    max_score = max(grades)
    min_score = min(grades)
    count_over_or_equal_avg = sum(1 for g in grades if g >= average)

    return average, max_score, min_score, count_over_or_equal_avg


def main() -> None:
    """
    サンプルの成績データを分析し、結果を表示する。
    """
    datasets = [
        [85, 90, 78, 92, 88],
        [70, 75, 80, 85, 90],
    ]

    for grades in datasets:
        avg, max_s, min_s, count = analyze_grades(grades)

        print(f"平均点: {avg:.1f}")
        print(f"最高点: {max_s}")
        print(f"最低点: {min_s}")
        print(f"平均以上の学生数: {count}")
        print("")


if __name__ == "__main__":
    main()
```
